# Visualisasi Data

```python
import matplotlib.pyplot as plt

#Pie chart labels
label_list = data['KnownSex'].value_counts(sort=False).index
plt.axis("equal") #The pie chart is oval by default. To make it a circle use pyplot.axis("equal")

#To show the percentage of each pie slice, pass an output format to the autopctparameter 
plt.pie(data['KnownSex'].value_counts(sort=False),labels=label_list,autopct="%1.1f%%") 
plt.title("Gender of Blue Jay Birds")
plt.show()
```
