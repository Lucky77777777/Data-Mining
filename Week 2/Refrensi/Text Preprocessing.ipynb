# Library

```python
# Common Library
import pandas as pd
import numpy as np

# Library for text preprocessing
import re
import nltk
from Sastrawi.StopWordRemover.StopWordRemoverFactory import StopWordRemoverFactory, ArrayDictionary, StopWordRemover
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory
```

# Load Dataset

```python
data = pd.read_excel('D:\Clone Repo\Introduction-to-Data-Mining-Course-Material\Week 2\data teks preproces.xlsx',header=0)
data.head()
```

# Cleaning "judul" column 

```python
# Preprocessing
x1 = data.judul

documents = [] # Make empty list

for i in range(0, len(x1)):
    # Remove URL
    text = re.sub('((www\.[^\s]+)|(https?://[^\s]+))','', str(x1[i]))
    text = re.sub(r'#([^\s]+)', '', text)

    # Remove all the special characters
    document = re.sub(r'\W', ' ', text)
    
    # remove all single characters
    document1 = re.sub(r'\s+[a-zA-Z]\s+', ' ', document)
    
    # Remove single characters from the start
    document2 = re.sub(r'\^[a-zA-Z]\s+', ' ', document1) 
    
    # Remove numeric
    document3 = re.sub(r"\d+", "", document2)
    
    # Substituting multiple spaces with single space
    document4 = re.sub(r'\s+', ' ', document3, flags=re.I)
    
    # Converting to Lowercase
    document5 = document4.lower()

    # Remove whitespace at begin and end string
    document6 = document5.strip()
    
    documents.append(document6)

x1_v1 = documents
x1_v1
```

## Stopword Removal "judul" Column

```python
stop_factory = StopWordRemoverFactory().get_stop_words() # Get Stopword
stop_factory[1:6] # Call 5 examples stopword
```

```python
len(stop_factory)
```

```python
# Adding more stopword from file 
more_stopword = open("D:\Clone Repo\Introduction-to-Data-Mining-Course-Material\Week 2\Stopword_list.txt", "r")
more_stopword = more_stopword.read()
stopword_list = more_stopword.split("\n")
```

```python
# Combine it
Stopword = stop_factory + stopword_list
dictionary = ArrayDictionary(Stopword)
```

```python
STR = StopWordRemover(dictionary)
doc_clean = [] # Make empty lits

for i in range(0, len(x1_v1)):
    kalimat = STR.remove(x1_v1[i])
    doc_clean.append(kalimat)
```

```python
doc_clean
```

## Stemming 'judul' column 

```python
factory = StemmerFactory()
stemmer = factory.create_stemmer()
judul_clean = [stemmer.stem(word) for word in doc_clean]
```

# Export

```python
data['judul clean']=judul_clean
```
